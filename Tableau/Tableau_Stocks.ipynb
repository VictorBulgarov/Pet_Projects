```python
import pandas as pd
```

```python
Apple = pd.read_csv("AAPL.csv") # Используем данные об акциях с сайта https://finance.yahoo.com
Amazon = pd.read_csv("AMZN.csv")
Google = pd.read_csv("GOOG.csv")
Netflix = pd.read_csv("NFLX.csv")
Tesla = pd.read_csv("TSLA.csv")
Sony = pd.read_csv("SONY.csv")
```

```python
df_all = [Apple, Amazon, Google, Netflix, Tesla, Sony]# Делаем общий список данных
```

```python
for df in df_all:
    df['MA50'] = df.Close.rolling(50).mean()
    df['MA200'] = df.Close.rolling(200).mean() # Создаем скользящую среднюю 
    #(Скользящее среднее может помочь выявить общие тренды в данных. 
    #Например, если 50-дневное скользящее среднее пересекает 200-дневное 
    #скользящее среднее снизу вверх, это может считаться бычьим (положительным) сигналом.)
```

```python
Apple.head(200)
```

```python
for df in df_all:
    df["Previous day close price"] = df.Close.shift(1)
    # Создаем столбец "Цена закрытия предыдущего дня"
```

```python
Apple.head(200)
```

```python
for df in df_all:
    df["Change in price"] = df["Close"] - df["Previous day close price"]
    # Создаем столбец "Изменение цены"
```

```python
Apple.head()
```

```python
for df in df_all:
    df["Percent сhange in price"] = df.Close.pct_change()
    # Создаем столбец "Процентное изменение цены"
```

```python
Apple.head()
```

```python
for df in df_all:
    df["Previous day volume"] = df.Volume.shift(1)
    # Создаем столбец "Объем предыдущего дня"
```

```python
Apple.head()
```

```python
for df in df_all:
    df["Change in volume"] = df["Volume"] - df["Previous day volume"]
    # Создаем столбец "Изменение объема"
```

```python
for df in df_all:
    df["Percent change in volume"] = df.Volume.pct_change()
    # Создаем столбец "Процентное изменение объема"
```
